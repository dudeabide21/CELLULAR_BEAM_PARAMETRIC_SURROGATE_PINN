```python
# pip install pyvista h5py numpy

import h5py
import numpy as np
import pyvista as pv


def infer_celltype(K: int) -> int:
    if K == 4: return pv.CellType.TETRA
    if K == 8: return pv.CellType.HEXAHEDRON
    if K == 6: return pv.CellType.WEDGE
    if K == 5: return pv.CellType.PYRAMID
    raise ValueError(f"Unsupported cell size K={K}. Add mapping for your element type.")


def load_mesh(path="simulation_data.h5", group="group_01"):
    with h5py.File(path, "r") as f:
        g = f[group]
        points = np.asarray(g["mesh_pos"][:], dtype=np.float64)
        cells = np.asarray(g["cells"][:], dtype=np.int64)
        stress = np.asarray(g["eq_stress"][:]).ravel()
        deform = np.asarray(g["deformation"][:]) if "deformation" in g else None

    if points.ndim != 2 or points.shape[1] != 3:
        raise ValueError(f"mesh_pos must be (N,3). Got {points.shape}")

    N = points.shape[0]

    # Strip leading count column if present: [K, n0, n1, ...]
    if cells.ndim != 2:
        raise ValueError(f"cells must be 2D (M,K). Got {cells.shape}")
    if cells.shape[1] >= 2:
        first = cells[:, 0]
        if np.all(first == first[0]) and first[0] == (cells.shape[1] - 1):
            cells = cells[:, 1:]

    M, K = cells.shape

    # Fix 1-based indexing if needed
    cmin, cmax = int(cells.min()), int(cells.max())
    if cmin >= 1 and cmax <= N and 0 not in cells:
        cells = cells - 1

    cmin, cmax = int(cells.min()), int(cells.max())
    if cmin < 0 or cmax >= N:
        raise ValueError(
            f"Connectivity indices invalid after fixes: min={cmin}, max={cmax}, N={N}. "
            "cells encoding likely differs (mixed types/offsets)."
        )

    celltype = infer_celltype(K)
    cells_vtk = np.hstack([np.full((M, 1), K, dtype=np.int64), cells]).ravel()
    celltypes = np.full(M, celltype, dtype=np.uint8)
    mesh = pv.UnstructuredGrid(cells_vtk, celltypes, points)

    # Attach stress properly
    if stress.shape[0] == N:
        mesh.point_data["eq_stress"] = stress
    elif stress.shape[0] == M:
        mesh.cell_data["eq_stress"] = stress
        mesh = mesh.cell_data_to_point_data()  # smooth colors
    else:
        raise ValueError(f"eq_stress length {stress.shape[0]} doesn't match nodes {N} or cells {M}")

    # Optional deformation
    if deform is not None and deform.ndim == 2 and deform.shape[0] == N and deform.shape[1] >= 3:
        mesh.point_data["deformation"] = deform[:, :3].astype(np.float64)

    return mesh


def show(mesh, group="group_01", deform_scale=0.0):
    m = mesh.copy(deep=True)
    if deform_scale and "deformation" in m.point_data:
        m.points = m.points + deform_scale * m.point_data["deformation"]

    p = pv.Plotter(window_size=(1700, 1000))
    p.enable_anti_aliasing("ssaa")
    p.set_background("white")

    # Main mesh: LESS smooth, EDGES ON
    p.add_mesh(
        m,
        scalars="eq_stress",
        smooth_shading=False,      # ⬅️ key change
        lighting=True,
        show_edges=True,           # ⬅️ bring back mesh lines
        edge_color="black",
        line_width=0.8,
        ambient=0.25,
        diffuse=0.75,
        specular=0.05,
        scalar_bar_args=dict(
            title="eq_stress",
            vertical=True,
            position_x=0.88,
            position_y=0.12,
            height=0.76,
            label_font_size=12,
            title_font_size=14,
        ),
    )

    # Optional: subtle wireframe overlay (helps with depth perception)
    p.add_mesh(
        m,
        style="wireframe",
        color="black",
        opacity=0.15,
    )

    # Orientation helpers
    p.add_axes(line_width=2)
    p.add_bounding_box(color="black")

    # Clean, non-overlapping title
    p.add_text(f"{group} | eq_stress", position="upper_left", font_size=14, color="black")

    # Camera sanity
    p.camera_position = "iso"
    p.reset_camera()
    p.camera.zoom(1.15)

    p.show()



if __name__ == "__main__":
    group = "group_01"
    mesh = load_mesh("../Data/simulation_data.h5", group=group)

    # deform_scale=0 means undeformed; try 5/10/20 if you want
    show(mesh, group=group, deform_scale=0.0)
```

```python
#normalizing the parameters across different groups

class MinMaxToMinusOneOne:
    def __init__(self, X):
        self.min = X.min(axis=0)
        self.max = X.max(axis=0)
        self.scale = self.max - self.min
        self.scale[self.scale == 0] = 1.0

    def encode(self, X):
        return 2.0 * (X - self.min) / self.scale - 1.0

class ScaleByMaxAbs:
    def __init__(self, U):
        self.s = np.max(np.abs(U))
        if self.s == 0: self.s = 1.0
    def encode(self, U):
        return U / self.s
    def decode(self, Uhat):
        return Uhat * self.s
```

```python
import torch
import torch.nn as nn

# gradient helper code

class MLP(nn.Module):
    def __init__(self, in_dim, out_dim=3, width=128, depth=5):
        super().__init__()
        layers = [nn.Linear(in_dim, width), nn.Tanh()]
        for _ in range(depth - 1):
            layers += [nn.Linear(width, width), nn.Tanh()]
        layers += [nn.Linear(width, out_dim)]
        self.net = nn.Sequential(*layers)

        for m in self.net:
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                nn.init.zeros_(m.bias)

    def forward(self, x):
        return self.net(x)
```

```python
import torch
from torch.autograd import grad

# residual function for linear elasticity

def elasticity_residual(x, u, E, nu):
    mu = E / (2.0 * (1.0 + nu))
    lam = E * nu / ((1.0 + nu) * (1.0 - 2.0 * nu))

    ux, uy, uz = u[:, 0:1], u[:, 1:2], u[:, 2:3]

    dux = grad(ux, x)
    duy = grad(uy, x)
    duz = grad(uz, x)

    exx = dux[:, 0:1]
    eyy = duy[:, 1:2]
    ezz = duz[:, 2:3]
    exy = 0.5 * (dux[:, 1:2] + duy[:, 0:1])
    exz = 0.5 * (dux[:, 2:3] + duz[:, 0:1])
    eyz = 0.5 * (duy[:, 2:3] + duz[:, 1:2])

    tr = exx + eyy + ezz

    sxx = lam * tr + 2 * mu * exx
    syy = lam * tr + 2 * mu * eyy
    szz = lam * tr + 2 * mu * ezz
    sxy = 2 * mu * exy
    sxz = 2 * mu * exz
    syz = 2 * mu * eyz

    dsxx = grad(sxx, x)
    dsxy = grad(sxy, x)
    dsxz = grad(sxz, x)
    dsyy = grad(syy, x)
    dsyz = grad(syz, x)
    dszz = grad(szz, x)

    rx = dsxx[:, 0:1] + dsxy[:, 1:2] + dsxz[:, 2:3]
    ry = dsxy[:, 0:1] + dsyy[:, 1:2] + dsyz[:, 2:3]
    rz = dsxz[:, 0:1] + dsyz[:, 1:2] + dszz[:, 2:3]

    return torch.cat([rx, ry, rz], dim=1)
```

```python
#computing loss for PINN

def pinn_losses(model, x, u_true, E, nu, w_data=1.0, w_pde=1e-6):
    u_pred = model(x)
    loss_data = torch.mean((u_pred - u_true)**2)

    r = elasticity_residual(x, u_pred, E, nu)
    loss_pde = torch.mean(r**2)

    loss = w_data*loss_data + w_pde*loss_pde
    return loss, loss_data, loss_pde

```
